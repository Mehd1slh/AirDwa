```python
import folium
from folium.plugins import Draw

# Center the map over the High Atlas Mountains (Morocco)
m = folium.Map(location=[31.1, -7.9], zoom_start=9)

# Add the Draw tool so you can draw a Bounding Box
draw = Draw(
    draw_options={'polyline': False, 'polygon': False, 'circle': False, 'marker': False, 'circlemarker': False},
    edit_options={'edit': False}
)
m.add_child(draw)

# INSTEAD of display(m), save it as an HTML file!
m.save("bbox_selector.html")
print("✅ Map saved! Open 'bbox_selector.html' in your web browser.")
```
